# Алгоритмы и структуры данных: Тема 1
## Анализ сложности алгоритмов ($O$-нотация)

Сегодня мы разберем, как оценивать эффективность кода не по секундомеру (который зависит от железа), а по количеству операций в зависимости от объема входных данных.

### Почему это важно?
Представьте, что у вас есть массив из 10 элементов и алгоритм с $O(n^2)$. Это 100 операций. Если элементов станет 1 000 000, операций будет 1 000 000 000 000. Ваша программа просто зависнет. 

--- 
## 1. Константное время: $O(1)$

Алгоритм выполняется за одно и то же время независимо от объема данных. 
**Пример:** Доступ к элементу массива по индексу.

````kotlin
fun getFirstElement(list: List<Int>): Int? {
    return if (list.isNotEmpty()) list[0] else null
}

val data = listOf(10, 20, 30, 40, 50)
println("Первый элемент: ${getFirstElement(data)}")
````

## 2. Линейное время: $O(n)$

Количество операций растет прямо пропорционально количеству элементов. Если данных в 2 раза больше — времени нужно в 2 раза больше.
**Пример:** Поиск элемента в неотсортированном списке.

 Мы проверяем каждый элемент по одному разу. Если в массиве $n$ элементов, нам нужно максимум $n$ шагов.

````kotlin
fun findElement(list: List<Int>, target: Int): Int? {
    for (index in list.indices) {
        if (list[index] == target) return index
    }
    return -1
}

val target = 40
println("Индекс элемента $target: ${findElement(data, target)}")
````

## 3. Квадратичное время: $O(n^2)$

Обычно встречается во вложенных циклах. Время выполнения растет как квадрат от входных данных.
**Пример:** Поиск дубликатов простым перебором.

Обычно это признак того, что код будет тормозить на больших данных. Вложенные циклы — это классика $O(n^2)$.

````kotlin
fun hasDuplicates(list: List<Int>): Boolean {
    for (i in list.indices) {
        for (j in i + 1 until list.size) {
            if (list[i] == list[j]) return true
        }
    }
    return false
}

val listWithDups = listOf(1, 2, 3, 2)
println("Есть дубликаты: ${hasDuplicates(listWithDups)}")
````

````kotlin
fun bubbleSort(arr: IntArray) {
    val n = arr.size
    for (i in 0 until n - 1) {
        for (j in 0 until n - i - 1) {
            if (arr[j] > arr[j + 1]) {
                // Swap
                val temp = arr[j]
                arr[j] = arr[j + 1]
                arr[j + 1] = temp
            }
        }
    }
}
````

## 4. Логарифмическое время: $O(\log n)$

Это «золотой стандарт» эффективности для поиска. На каждом шаге мы отсекаем половину данных.
**Пример:** Бинарный поиск в отсортированном массиве.

````kotlin
fun binarySearch(list: List<Int>, target: Int): Int {
    var low = 0
    var high = list.size - 1

    while (low <= high) {
        val mid = (low + high) / 2
        val guess = list[mid]
        if (guess == target) return mid
        if (guess > target) high = mid - 1
        else low = mid + 1
    }
    return -1
}

val sortedData = listOf(10, 20, 30, 40, 50, 60, 70)
println("Бинарный поиск (40): ${binarySearch(sortedData, 40)}")
````

## Квадратичная сложность: $O(n^2)$

Обычно возникает, когда внутри одного цикла находится другой. Типичный пример — поиск дубликатов в массиве путем сравнения каждого элемента с каждым.

````kotlin
fun hasDuplicates(arr: IntArray): Boolean {
    val n = arr.size
    for (i in 0 until n) {
        for (j in i + 1 until n) {
            if (arr[i] == arr[j]) return true
        }
    }
    return false
}
````

# Экспоненциальная (степенная) сложность: $O(2^n)$

Это «кошмар» оптимизатора. Время выполнения удваивается при добавлении всего одного элемента. Классика — наивное рекурсивное вычисление чисел Фибоначчи.

````kotlin
fun fib(n: Int): Int {
    if (n <= 1) return n
    return fib(n - 1) + fib(n - 2)
}
````

**Блок-схема (Mermaid):**
Здесь дерево вызовов растет в геометрической прогрессии. Для простоты покажем логику одного вызова:

### Краткое сравнение
| Сложность | Что это значит | Пример |
| :--- | :--- | :--- |
| **$O(n^2)$** | Увеличили данные в 10 раз — время выросло в 100 раз. | Вложенные циклы, простые сортировки. |
| **$O(2^n)$** | Добавили 1 элемент — время выросло в 2 раза. | Рекурсивный перебор, взлом паролей «в лоб». |

> **Забавный факт:** При $O(2^n)$, если для $n=40$ программа работает пару минут, то для $n=100$ она может работать дольше, чем существует Вселенная.

## Итоговая шпаргалка сложности

| Обозначение | Название | Пример |
| :--- | :--- | :--- |
| $O(1)$ | Константная | Чтение из массива по индексу |
| $O(\log n)$ | Логарифмическая | Бинарный поиск |
| $O(n)$ | Линейная | Проход циклом по списку |
| $O(n \log n)$ | Линеаритмическая | Быстрая сортировка (Merge Sort) |
| $O(n^2)$ | Квадратичная | Вложенные циклы |
| $O(2^n)$ | Экспоненциальная | Рекурсивный поиск чисел Фибоначчи |
